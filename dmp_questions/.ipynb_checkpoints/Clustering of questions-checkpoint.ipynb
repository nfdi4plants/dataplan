```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import umap
import hdbscan
import numpy as np
import plotly.express as px
from sklearn.cluster import KMeans
np.random.seed(11)
import umap.plot
# model = SentenceTransformer('distilbert-base-nli-mean-tokens')
s_words= ["not","who","long","are","also","there","use","especially","has","as","ended", "their","you","re","after",
           "for","research","any","existing","should","end", "shared","and", "with","other", "suitable","potential",
         "or","by","once","new","in", "all","subsequent","if", "how", "what", "is", "way" ,"being", 
          "they", "your", "of", "adequate","do", "does","within","arise","allow", "which"]
with open("DFG.txt") as f:
    dfg = f.read()
dfg_q1 = dfg.split("\n")
dfg_q1 = [q for q in dfg_q1 if q != "" and q != " "]
with open ("Horizon_Europe.txt") as f:
    h_e = f.read()
h_e_q = h_e.split("\n")
h_e_q = [q for q in h_e_q if q != "" and q != " "]

dfg_q = dfg_q1+h_e_q

# embedder = SentenceTransformer('allenai-specter')
embedder = SentenceTransformer('stsb-mpnet-base-v2')
embeddings = embedder.encode(dfg_q)



umap_embeddings = umap.UMAP(n_neighbors= 3, metric='cosine',random_state=42, min_dist =0.1 ).fit_transform(embeddings)

np.random.seed(11)
cluster = hdbscan.HDBSCAN(min_cluster_size=3,
                          metric='euclidean',min_samples=5,                      
                          cluster_selection_method='eom',prediction_data=True, approx_min_span_tree=False).fit(umap_embeddings)

# cluster_model = KMeans(n_clusters=5)
# cluster = cluster_model.fit(umap_embeddings)
# Prepare data
umap_data = umap_embeddings
result = pd.DataFrame(umap_data, columns=['x', 'y'])
result['labels'] = cluster.labels_
outliers = result.loc[result.labels == -1, :]
clustered = result.loc[result.labels >= -1, :]

docs_df = pd.DataFrame(dfg_q, columns=["Doc"])
docs_df['Topic'] = cluster.labels_
docs_df['Doc_ID'] = range(len(docs_df))
docs_per_topic = docs_df.groupby(['Topic'], as_index = False).agg({'Doc': ' '.join})

import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

def c_tf_idf(documents, m, ngram_range=(1, 1)):
    count = CountVectorizer(ngram_range=ngram_range, stop_words=s_words).fit(documents)
    t = count.transform(documents).toarray()
    w = t.sum(axis=1)
    tf = np.divide(t.T, w)
    sum_t = t.sum(axis=0)
    idf = np.log(np.divide(m, sum_t)).reshape(-1, 1)
    tf_idf = np.multiply(tf, idf)

    return tf_idf, count
  
tf_idf, count = c_tf_idf(docs_per_topic.Doc.values, m=len(dfg_q))

def extract_top_n_words_per_topic(tf_idf, count, docs_per_topic, n=20):
    words = count.get_feature_names_out()
    labels = list(docs_per_topic.Topic)
    tf_idf_transposed = tf_idf.T
    indices = tf_idf_transposed.argsort()[:, -n:]
    top_n_words = {label: [(words[j], tf_idf_transposed[i][j]) for j in indices[i]][::-1] for i, label in enumerate(labels)}
    return top_n_words

def extract_topic_sizes(df):
    topic_sizes = (df.groupby(['Topic'])
                     .Doc
                     .count()
                     .reset_index()
                     .rename({"Topic": "Topic", "Doc": "Size"}, axis='columns')
                     .sort_values("Size", ascending=False))
    return topic_sizes

top_n_words = extract_top_n_words_per_topic(tf_idf, count, docs_per_topic, n=20)
topic_sizes = extract_topic_sizes(docs_df); 
topic_sizes.head(10)
word2 = list(map(lambda x: x[0:3],top_n_words.values()))

clustered["Question"] = dfg_q
clustered["Word Count"] =list(map(lambda x: len(x.split(" ")), dfg_q))
clustered["Topics"] = list(map(lambda x: word2[int(x)][0][0]+ ", " +  word2[int(x)][1][0]+", " +  word2[int(x)][2][0], clustered["labels"]))
clustered1 =clustered.sort_values(by=['labels'])


fig = px.scatter(clustered1 , x="x", y="y", color='Topics' ,size="Word Count",
template="simple_white",  title="Clustering of the all questions in the DFG and Horizon Europe questions" )
fig.update_layout(
    xaxis_title=" ",
    yaxis_title=" ",
    font=dict(
        family="helvetica",
        size=12,
        color="black"
    )
)
j=0
# for i,y in enumerate(clustered1["y"]):
# # j= j+1 if  clustered1["labels"][i] 
#     fig.add_annotation(x=clustered1["x"][i], y=y+i,
#                 text=dfg_q[i],
#                 showarrow=True,
#                 arrowhead=1)

fig.show()
```
